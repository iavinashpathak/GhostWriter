```python
import pandas as pd
```

```python
df = pd.read_csv('articles.csv')
df.head()
```

```python
df['full_text'] = df['title'].astype(str) + ' ' + df['text'].astype(str)
```

```python
output_file = 'data/corpus.txt'

df['full_text'].to_csv(
    output_file, 
    index=False,
    header=False)
```
